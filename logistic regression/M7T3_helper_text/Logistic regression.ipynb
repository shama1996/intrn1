### Precison and Recall metrics

###### Precision and recall handle imbalanced dataset efficiently.
__precision__=prediction that are actually positive/total predicted positive.   (check formula in abobe image)
      
      - Lets understand precison with the help of an example:Consider you are a detective and task is catching a criminal,& you know thosa criminals are attending a party full of important people.
      you can imagine,wrongly declaring any of these important people as criminal can have dire consequences for your agency.So,as a detective you can only arrest a criminals.(To be perfectly clear) .
      
  __Avoiding the wrongful arrest of these VIPs take precedence over catching the criminals__.
  
In order to do that,we must avoid arresting  VIPs as criminals.ie __minimizing the false positive__.
    But in order to minimize false positive,high chances of real criminals escaping and not being caught .ie -__false negative increases__.
    In a nutshell precision is used when avoiding false positive is more essential than encountering false negative

##Importing Librarires

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action = 'ignore')
plt.style.use(['seaborn-bright', 'dark_background'])
```

## Importing dataset

```python
data = pd.read_csv('churn_prediction_simple.csv')
data.head()
```

```python
data = data.dropna()
data.info()
```

```python
# checking the dataset distribution
data['churn'].value_counts()/len(data)
```

```python
#separating dependent and independent varibales
X = data.drop(columns = ['churn', 'customer_id'])
Y = data['churn']
```

```python
#Scaling the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(scaled_X, Y, train_size = 0.80, stratify = Y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Model Building, predictions and odds ratio

```python
from sklearn.linear_model import LogisticRegression as LR
classifier = LR(class_weight = 'balanced')
```

```python
classifier.fit(x_train, y_train)
predicted_values = classifier.predict(x_test)              # Predicting class
predicted_probabilities = classifier.predict_proba(x_test) # Predicting probabilities
```

```python
predicted_values
```

```python
predicted_probabilities, predicted_probabilities.shape
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test, predicted_values)
cf
```

##Evaluation Metrics

```python
#Accuracy
classifier.score(x_test, y_test)
```

```python
#calculating the precision score
from sklearn.metrics import precision_score
Precision = precision_score(y_test, predicted_values)
Precision
```

```python
#calculating recall score
from sklearn.metrics import recall_score
Recall = recall_score(y_test, predicted_values)
Recall
```

```python
# manually calculating the f1 score
f1 = 2/((1/Precision)+(1/Recall))
f1
```

```python
#calculating the f1-score
from sklearn.metrics import f1_score
F1 = f1_score(y_test, predicted_values)
F1
```

```python
# calculating precision, recall, f1-score and support at once.
from sklearn.metrics import precision_recall_fscore_support as PRF_summary
precision, recall, f1, support = PRF_summary(y_test, predicted_values)
```

```python
from sklearn.metrics import classification_report
k = classification_report(y_test, predicted_values)
print(k)
```

## Precision-Recall Curve

```python
# gathering Precision/recall scores for different thresholds
from sklearn.metrics import precision_recall_curve
precision_points, recall_points, threshold_points = precision_recall_curve(y_test, predicted_probabilities[:,1])
precision_points.shape, recall_points.shape, threshold_points.shape
```

```python
plt.figure(figsize = (7,5), dpi = 100)
plt.plot( threshold_points, precision_points[:-1], color = 'green', label = 'Precision Curve')
plt.plot( threshold_points, recall_points[:-1], color = 'orange', label = 'Recall Curve')
plt.xlabel('Threshold Points', fontsize = 15)
plt.ylabel('Score', fontsize = 15)
plt.title('Precision-Recall tradeoff', fontsize = 20)
plt.legend()
```

## AUC-ROC Curve

```python
from sklearn.metrics import roc_curve , roc_auc_score
fpr, tpr, threshold = roc_curve(y_test, predicted_probabilities[:,1])
```

```python
plt.figure(figsize = (7,5), dpi = 100)
plt.plot( fpr, tpr, color = 'green')
plt.plot( [0,1], [0,1], label = 'baseline', color = 'red')
plt.xlabel('FPR', fontsize = 15)
plt.ylabel('TPR', fontsize = 15)
plt.title('AUC-ROC', fontsize = 20)
plt.show()
roc_auc_score(y_test, predicted_probabilities[:,1])
```

## Coefficient Plot

```python
# arranging the data
c = classifier.coef_.reshape(-1)
x = X.columns

coeff_plot = pd.DataFrame( {
                            'coefficients': c,
                            'variable' : x
                            })

# sorting the values
coeff_plot = coeff_plot.sort_values(by = 'coefficients')
coeff_plot.head()
```

```python
plt.figure(figsize=(8, 6), dpi=120)
plt.barh( coeff_plot['variable'] , coeff_plot['coefficients'] )
plt.xlabel( "Coefficient Magnitude", fontsize = 15)
plt.ylabel('Variables', fontsize = 15)
plt.title('Coefficient plot', fontsize = 20)
```
