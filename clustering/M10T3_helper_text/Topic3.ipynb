### A marketing firm wants to run promotionla compaign n different parts of country,so in order to do that firm needs to understands diversity in the population demography,so that it can plan the promotional compaign accordingly.

We are provided with population data of different region along with demographic.

Objective: to segragate the regions into different groups,so that markettig teams can plan their resource accordingly

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
data = pd.read_csv("Population_Data.csv")
data.info()
```

```python
data.head()
```

```python
numeric = ["Indians","Foreigners",
           "Indian_Male","Indian_Female",
           "Foreigners_Female",
           'Foreigners_Male',"Total Population"]
```

```python
def cleaner(z):
  return z.replace(',','')
```

```python
cleaner('100,234')
```

```python
for i in data[numeric]:
  data[i] = data[i].apply(cleaner)

data.head()
```

```python
data.info()
```

althogh we have removed comma from the object,the datatype is still in object type.therefore we convert them to numeric type explicitely

```python
data[numeric] = data[numeric].apply(pd.to_numeric)
data.info()
```

##### verifying the integrity of data
it means we have column corresponding to population, indians & foreigner & also the total population.A simple integrity check would be to make sure that indian population + foreign population=total population.

For this we add up all the values in column imdians & foreigner & then again the 2 column ,& then subtracted from sum of total population

```python
data[['Indians','Foreigners']].sum().sum() - data['Total Population'].sum()
```

we get 0 as he outputs,it means indians & foreigner  add upto the total population

```python
#checking,whether these 4 columns add upto the total population
data[['Indian_Male','Indian_Female',"Foreigners_Male",'Foreigners_Female']].sum().sum() - data['Total Population'].sum()
```

we find that output is a negative value.Why is the value negative????
it means total population is greater in value than sum of all males & females.
It means there are people in the population who do not identify themselves as male or female.This is an important piece of information as it produce new knowledge in our modelling process.

lets create  a new feature called 'other' which consist of population who does'nt  identify themselevs as male or female

```python
MF_sum = data['Indian_Male'] + data['Indian_Female'] + data["Foreigners_Male"] + data['Foreigners_Female']
data['other'] = data['Total Population'] - MF_sum
data.head()
```

```python
data['Region'].nunique(),data['Office Location Id'].nunique()
```

* these feature donot provide useful information

* we also drop total population as it is already explained as the sum of indians & foreigner

```python
data1 = data.drop(columns = ['Region',"Office Location Id", 'Total Population'])
data1.head()
```

```python
#normalizing all the variables
from sklearn.preprocessing import Normalizer
norm = Normalizer()
columns = data1.columns
data1 = norm.fit_transform(data1)

data1 = pd.DataFrame(data1, columns = columns)
data1.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(data1)
pred=kmeans.predict(data1)
```

```python
pred,len(pred)
```

```python
pd.Series(pred).value_counts()
```

```python
kmeans.inertia_
```

how do we know ,if '2' is the right no. of clusters??


we can find out by plotting the graph between no.of cluster & inertia

```python
kmeans.score(data1)
```

```python
SSE = [] #the empty list contains inertia score,whenever a clustering process is perfored over the data
for cluster in range(1,10):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(data1)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster':range(1,10), 'SSE':SSE})
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data1)
pred=kmeans.predict(data1)
```

```python
data1['cluster'] = pred
```

```python
def seg(str_x, str_y, clusters):
  x = []
  y = []

  for i in range(clusters):
    x.append(data1[str_x][data1['cluster']==i])
    y.append(data1[str_y][data1['cluster']==i])
  
  return x,y

def plot_clusters(str_x, str_y, clusters):
  plt.figure(figsize = (5,5), dpi = 120)
  
  x,y = seg(str_x, str_y, clusters)

  for i in range(clusters):
    plt.scatter(x[i], y[i], label = 'cluster{}'.format(i))
  
  plt.xlabel(str_x)
  plt.ylabel(str_y)
  plt.title(str(str_x+" Vs "+str_y))
  plt.legend()
```

```python
plot_clusters('Indians', 'Foreigners',3)
```

```python
plot_clusters('Indian_Male', 'Foreigners_Male',3)
```

```python
plot_clusters('Indian_Female', 'Foreigners_Female',3)
```

```python
plot_clusters('other', 'Indians',3)
```

```python
[plot_clusters(x,y,3) for x in data1.drop(columns=['cluster']).columns for y in data1.drop(columns=['cluster']).columns if x!=y]
```
