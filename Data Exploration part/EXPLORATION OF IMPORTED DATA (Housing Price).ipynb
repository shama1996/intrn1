#### Diffrent ways of treating outliers

- 1. DELETION
- 2. CAPPING OR IMPUTING
- 3. DATA TRANSFORMATION
- 4. BINNIG

- __Deletion__ is least preffered
- In __Capping__ outliers is not removed but is replaced by mean,median or mode whichever makes least impact to the variable
- In __DataTransformation__ the variable is transformed to its log value or square or cube root value since log/square/cube roots of a outliers is likely to in range of log or square/cube root of rest of the data points.
- In __Binning__ we formed different bins based on value of variable to treat the outliers

### Treating outliers by imputing

```python
q1=data['Sale Price'].quantile(0.25)
q2=data['Sale Price'].quantile(0.75)
```

```python
iqr=q2-q1
iqr
```

```python
upper_limit=q2+1.5*iqr
lower_limit=q1-1.5*iqr
upper_limit,lower_limit
```

```python
def limit_imputer(value):
    if  value>upper_limit:
        return upper_limit
    if value<lower_limit:
        return lower_limit
    else:
        return value
```

```python
data['Sale Price']=data['Sale Price'].apply(limit_imputer)
```

```python
#again checking descritive statistics and comparing from above
data['Sale Price'].describe()
```

- We observe that maxm value has changed to the upper value,means that there are no more value beyond upper limit in our dataset

# Feature Scaling
 - When the model is built,variables with higher magnitude and rane will have higher impact compared to the onces which are smaller.
 - The model might not predict properly since it is not giving equal weight to both the variables
  - __Feature Scaling__  transform all independent variable into similar magnitude and range so that model is not biased towars a particular variable
  
  Different techniques for feature scaling are:
   - __Standardsation__ It rescales the feature values so that they have properties of standard normal distribution with means as 0 & std deviation as 1
   - __Min Max Scaler__ The value range for transformed variable les between [0,1]
   - __Normalization__ The range is fixed from [-1,1]

```python
def slope_Error( slope, intercept ,  sample_data):
    sale = []
    for i in range(len(sample_data.flat_area)):
        tmp = sample_data.flat_area[i] * slope + intercept
        sale.append(tmp)
    MSE = mse(sample_data.sale_price, sale)
    return MSE
```

```python
slope = [i/10 for i in range(0,5000)]
Cost = []
for i in slope:
    cost = slope_Error( slope = i, intercept = 0, sample_data = sample_data)
    Cost.append(cost)
```

### gradient descent
 we will apply gradient descent to the sample data only to keep things simple

```python
#step 1:Initialize parameters-This functions take only the target variable as its parameter
def param_init(Y): 
  '''
  Y = target variable
  returns initialized values of m and c
  '''
  m = 0.1  # in this func we simply initiaze m as a small value 0.1 & c as the mean of the target variable and then return them
  c = Y.mean()
  return m, c
```

```python
#step 2:Generate predictions-In this we will define a func generate_predictions which will take slope m,intercept c & independent variable x
def generate_predictions(m, c, X):
  '''
  X = independent variable
  returns prediction generated by line with parameters m, c
  '''
  prediction = []
  for x in X:
    pred = (m * x) + c #this func returns a list containg predictions of every value in x w.r.t to m & c using eqn y=mx+c
    prediction.append(pred)
  return prediction
```

```python
#step 3:calculate the cost-for this we define a fun compute_cost,as we know cost is nothing but MSE b/w the predicted & the actuals
def compute_cost(prediction, Y):
  '''
  returns the mean_squared_error between prediction and Y
  '''
    cost = np.sum(((prediction - Y)**2)/len(Y))
    return cost
```

```python
#step 4:update parameters-This step is carried out in  2 parts 1; calcultating the gradients
                                                              # 2;using the gradients to update the parameters m & c
def gradients(prediction, Y, X):
  '''
  Returns Gradients corresponding to m and c
  '''
  n = len(Y)
    Gm = 2/n * np.sum((prediction - Y) * X)
    Gc = 2/n * np.sum((prediction - Y))
    return Gm, Gc
```

```python
def param_update(m_old , c_old , Gm_old, Gc_old, alpha):
  '''
  Update and return the new values of m and c
  '''
  m_new = m_old - alpha * Gm_old
  c_new = c_old - alpha * Gc_old
  return m_new, c_new
```

```python
def result(m, c, X, Y, cost, predictions, i):
    '''
    Print and plot the final result obtained from gradient descent
    '''
  ##If the Gradient Descent converged to the optimum value before max_iter
    if i < max_iter - 1 :
        print("***** Gradient Descent has converged at iteration {} *****".format(i))
    else:
        print("***** Result after", max_iter,'iterations is: *****')
  
  ## plotting the final result
    plt.figure(figsize = (14,7), dpi = 120)
    plt.scatter(X, Y, color = 'red', label = 'data points')
    label = 'final regression line: m = {}; c = {}'.format(str(m), str(c))
    plt.plot(X, predictions, color = 'green', label = label)
    plt.xlabel('flat_area')
    plt.ylabel('sale_price')
    plt.title('final regression line')
    plt.legend()
```

```python
## defining and reshaping the dataset
sale_price = sample_data['sale_price'].values.reshape(-1,1)
flat_area = sample_data['flat_area'].values.reshape(-1,1)

## declaring parameters
max_iter = 1000
cost_old = 0
alpha = 0.01

## Step 1: inititialising the values of m, c
m , c = param_init(sale_price)

## Gradient Descent in action
for i in range(0 , max_iter):
  
  ## Step 2: genrating predictions
  predictions = generate_predictions(m , c , flat_area)

  ## Step 3: calculating cost
  cost_new = compute_cost(predictions , sale_price)

  ## checking if GD converged
  if abs(cost_new - cost_old) < 10**(-7):
    break
  
  ## calculating gradients
  Gm, Gc = gradients(predictions, sale_price,
                     flat_area)
  
  ## Step 4: updating parameters m and c
  m, c = param_update(m, c, Gm, Gc, alpha)

  ## display result after every 20 iterations
  if i%20 == 0:
    print('After Iteration', i,': m =', m, '; c =', c,'; Cost =', cost_new)

  ##updating cost_old
  cost_old = cost_new

## final results
result(m , c, flat_area, sale_price,
       cost_new, predictions, i)
```
